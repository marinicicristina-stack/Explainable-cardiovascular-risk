```python
import pandas as pd

# Noms des colonnes du dataset Cleveland
columns = [
    "age", "sex", "cp", "trestbps", "chol", "fbs",
    "restecg", "thalach", "exang", "oldpeak",
    "slope", "ca", "thal", "target"
]

# Charger le dataset
df = pd.read_csv("heart_disease_cleveland.csv", names=columns)

# Afficher les 5 premières lignes
df.head()
```

```python
# Vérifier la taille du dataset
print("Nombre de patients :", df.shape[0])
print("Nombre de colonnes :", df.shape[1])

# Vérifier les types de données
print("\nTypes de données :")
print(df.dtypes)

# Vérifier les valeurs de la variable cible
print("\nValeurs de la cible :")
print(df["target"].value_counts().sort_index())
```

```python
# Rechercher les valeurs manquantes représentées par "?"
print("Nombre de '?' par colonne :")
print((df == "?").sum())
```

```python
import numpy as np

# Remplacer les "?" par NaN
df = df.replace("?", np.nan)

# Convertir ca et thal en valeurs numériques
df["ca"] = pd.to_numeric(df["ca"])
df["thal"] = pd.to_numeric(df["thal"])

# Vérifier les valeurs manquantes
print("Valeurs manquantes par colonne :")
print(df.isnull().sum())

print("\nNombre total de valeurs manquantes :", df.isnull().sum().sum())
```

```python
# Supprimer les lignes contenant des valeurs manquantes
df_clean = df.dropna().copy()

# Vérifier le résultat
print("Nombre de patients avant nettoyage :", len(df))
print("Nombre de patients après nettoyage :", len(df_clean))
print("Valeurs manquantes restantes :", df_clean.isnull().sum().sum())
```

```python
# Transformer la cible en classification binaire
# 0 = absence de maladie cardiaque
# 1 = présence de maladie cardiaque

df_clean["target"] = (df_clean["target"] > 0).astype(int)

# Afficher la distribution de la nouvelle cible
print("Distribution de la cible :")
print(df_clean["target"].value_counts())

print("\nDistribution en pourcentage :")
print(df_clean["target"].value_counts(normalize=True) * 100)
```
